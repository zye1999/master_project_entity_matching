```python
def save_to_file(file_name, contents):
    fh = open(file_name, 'a')
    fh.write(contents)
    fh.close()
```

```python
abt_buy_train.iloc[10,1:3]
```

```python
for i in range(len(lis)):
    print(abt_buy_train.iloc[lis[i],:])
    print("{:-^50s}".format("Split Line"))
    save_to_file('mobiles.txt', (abt_buy_train.iloc[lis[i],:]).to_string())
    save_to_file('mobiles.txt', '--------------------Split Line--------------------')
```
