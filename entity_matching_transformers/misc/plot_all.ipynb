### extract_value function

```python
import re
import numpy as np

def extract_value(filepath):

    f1_pattern = re.compile(r"f1_score: 0.[0-9]*")
    accuracy_pattern = re.compile(r"eval_loss: 0.[0-9]*")
    eval_pattern = re.compile(r"eval_loss: 0.[0-9]*")

    eval_results = open(filepath, "rt")
    contents = eval_results.read()
    eval_results.close()

    # a list of strings, containing data for each epoch
    f1_string_list = []
    accuracy_string_list = []
    eval_string_list = []

    f1_string_list = re.findall(f1_pattern, contents)
    accuracy_string_list = re.findall(accuracy_pattern, contents)
    eval_string_list = re.findall(eval_pattern, contents)

    f1_list = []
    accuracy_list = []
    eval_list = []

    # extract the float value from the each string
    for f1 in f1_string_list:
        f1_list.append(re.findall(r'0.[0-9]*', f1)[0])
        # f1_list = re.findall(r'0.[0-9]*', f1)
    f1_list = np.array(f1_list, dtype=float)
    
    for accuracy in accuracy_string_list:
        accuracy_list.append(re.findall(r'0.[0-9]*', accuracy))
    accuracy_list = np.array(accuracy_list, dtype=float)
        
    for eval in eval_string_list:
        eval_list.append(re.findall(r'0.[0-9]*', eval))
    eval_list = np.array(eval_list, dtype=float)
    
    return f1_list, accuracy_list, eval_list
```

### plot function

```python
import matplotlib.pyplot as plt
def plot_f1(dataset_name, path_bert, path_distilbert, path_roberta):
    f1_score_bert, _, _ = extract_value(path_bert)
    f1_score_roberta, _, _ = extract_value(path_roberta)
    f1_score_distilbert, _, _ = extract_value(path_distilbert)


    plt.plot(f1_score_bert, label='BERT')
    plt.plot(f1_score_roberta, label='RoBERTa')
    plt.plot(f1_score_distilbert, label='DistillBERT')
    # set y-limits of the current axes
    plt.ylim(0.6, 1)
    plt.xlabel('Number of epochs')
    plt.ylabel('F1')
    plt.title("Dataset: " + dataset_name)
    plt.legend(bbox_to_anchor=(1.05, 0.6))
    plt.show()
```
