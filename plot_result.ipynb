```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = ['overall coefficient', 'jaccard coefficient', 'dice coefficient']
y_data = [0.7563143124415341, 0.7494766224703419, 0.7542637521018496]
for i in range(len(x_data)):
    plt.bar(x_data[i], y_data[i])
plt.title("f1 score of merged all dataset")
plt.xlabel("method")
plt.ylabel("f1 score")
plt.show()
```

```python
all = [0.7563143124415341, 0.7494766224703419, 0.7542637521018496]
clean = [0.4815618221258134, 0.381562099871959, 0.38999999999999996]
dirty = [0.82527664531159, 0.8253414705027607, 0.8367171861595654]
x = np.arange(3)
total_width, n = 0.8, 3
width = total_width/n
x = x - (total_width - width)/2
plt.bar(x, all, width = width, label = "all")
plt.bar(x+width, clean, width = width, label = "clean")
plt.bar(x+2*width, dirty, width = width, label = "dirty")
plt.title("f1 score of token based method")
plt.xlabel("method")
plt.ylabel("f1 score")
plt.legend()
x_labels = ['overall coefficient', 'jaccard coefficient', 'dice coefficient']
plt.xticks(x+width, x_labels)
for i, j in zip(x,all):
    plt.text(i, j+0.01, "%.2f"%j, ha='center', va='bottom', fontsize=7)
for i, j in zip(x+width,clean):
    plt.text(i, j+0.01, "%.2f"%j, ha='center', va='bottom', fontsize=7)
for i, j in zip(x+2*width,dirty):
    plt.text(i, j+0.01, "%.2f"%j, ha='center', va='bottom', fontsize=7)
plt.show()
```
